```python
import sys
sys.path.append("/Library/Python/3.8/site-packages")

from datetime import datetime
from itertools import dropwhile, takewhile
from instagramy import *
import instaloader
import pandas as pd
import time
```

```python
# period to scrape the data from
SINCE = datetime(2022, 1, 1)
UNTIL = datetime(2022, 1, 15)

# users to scrape data from
USERS = ['googleplay', 'googlestudents']

NUM_BREAK = 250    # number of posts to fetch before sleeping
BREAK_TIME = 35    # amount of time (ms) to sleep
```

```python
# demo IG account

# username = "scraping_data_"
# password = "MoSscraping1"

# username = "I_scrape_data"
# password = "scraper1234"

username = "the_scraper_man"
password = "Scraping1234"

L = instaloader.Instaloader()
L.login(username, password)  # (login)
```

```python
# scrapes for all "users" between "since" and "until"

def get_likes_users(users, since, until):
    print("SINCE: " + str(since) + "\t UNTIL: "+ str(until))

    all_likes = {}

    for user in users:

        print("LIKES FOR: " + user )

        profile = instaloader.Profile.from_username(L.context, user)
        
        likes = get_likes(profile, since, until)

        all_likes[user] = likes
        
    return all_likes
```

```python
# scrapes data for a single "profile" between "since" and "until"

def get_likes(profile, since, until):
    likes = {}
    count = 0

    posts = profile.get_posts()
    
    for post in takewhile(lambda p: p.date > since, dropwhile(lambda p: p.date > until, profile.get_posts())):
    
            print("\tDATE: " + str(post.date))
#             print("\tCOUNT: " + str(count))
            print("")

            for likee in post.get_likes():
                count += 1
                
                if (count % NUM_BREAK == 0):
                    time.sleep(BREAK_TIME)
    
                if(likee in likes):
                    likes[likee] += 1

                else:
                    likes[likee] = 1
                    
    return likes
```

```python
# pretty prints the results

def pprint(users):
    
    print("\nRESULTS: ")

    for u in users:
        
        print("\n" + u + ": ")

        likes = dict(sorted(users[u].items(), key=lambda item: item[1], reverse=True))

        for likee in likes:
            print("\t" + likee.username + ": " + str(likes[likee]))
```

```python
def main():
    likes = get_likes_users(USERS, SINCE, UNTIL)
    pprint(likes)
```

```python
if __name__ == '__main__': 
    main()
```
